```python
# loop
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt


N = 100000
time = 1000
pi = np.zeros(time)

for i in range(time):
    x = uniform.rvs(-1, 2, N)
    y = uniform.rvs(-1, 2, N)
    d = np.sqrt(x**2 + y**2)
    # M = d < 1
    M = np.sum(d < 1)
    # print(4 * M / N)
    pi[i] = 4 * M / N

print(np.mean(pi))
print(np.std(pi))
plt.hist(pi, bins=30)
plt.show()
```

```python
# boardcast
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

time = 1000
N = 10000

X = uniform.rvs(-1, 2, (N, time))
Y = uniform.rvs(-1, 2, (N, time))
D = np.sqrt(X**2 + Y**2)
M = np.sum(D < 1, axis=0)
pi = 4 * M / N
print(np.mean(pi))
print(np.std(pi))
plt.hist(pi, bins=30)
plt.show()
```

```python
from scipy.stats import norm, ttest_ind 
from matplotlib import pyplot as plt

n = 10000
time = 1000
x1 = norm.rvs(0, 1, (n, time))
x2 = norm.rvs(0, 1, (n, time))
result = ttest_ind(x1, x2)
# print(result.pvalue)
plt.hist(result.pvalue, bins=30)    
plt.show()
```

```python
from scipy.stats import norm, ttest_ind
alpha = 0.05
N = 1000
n = 1000
X = norm.rvs(0, 1, (N, n))
Y = norm.rvs(0, 1, (N, n))
result = ttest_ind(X, Y, axis=1)
pvalue = result.pvalue
reject = pvalue < alpha
power = np.sum(reject) / N
print(power)
```
