```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt


n1,n2 = 40,30
N = 10000
X1 = norm.rvs(size=(N,n1))
X2 = norm.rvs(size=(N,n2))
S = np.mean(X1,axis=1) - np.mean(X2,axis=1)
# S = np.zeros(N)
# for i in range(N):
#     x1 = norm.rvs(size=n1)
#     x2 = norm.rvs(size=n2)
#     S[i] = np.var(x1)/np.var(x2)


plt.hist(S, bins=500, density=True, histtype='stepfilled', alpha=0.5)
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

N = 10000
n = 4000
x1 = chi2.rvs(df=1,size=(N,n))
s = np.mean(x1,axis=1)

plt.hist(s,bins=50,density=True,histtype='stepfilled',alpha=0.5)

plt.show()
```

```python
from scipy.stats import norm, chi2, uniform
import numpy as np

n = 20
N = 100000

x = uniform.rvs(loc=0, scale=1,size=(N,n))
s = x.mean(axis=1)

plt.hist(s,bins=50,density=True,histtype='stepfilled',alpha=0.5)
plt.show()    
```
