```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integral
 
# define anonymous function
f = lambda x : np.sin(x**2)
lb, ub = 0, 1
result = integral.quad(f, lb, ub)
print("The integration is {:.4f}".format(result[0]))
 
# draw the function line -------------------
x = np.linspace(0, np.pi/2, 100)
y = f(x)
fig, ax = plt.subplots(1, 1) 
ax.plot(x, y)
# draw the area ----------------------------
x = np.linspace(lb, ub, 100)
y = f(x)
# fill color between y and 0
ax.fill_between(x, y, 0, alpha=0.3, color='b')
ax.set_xlabel('X')
ax.set_title("The area = {:.4f}".format(result[0]))
plt.grid(True, linestyle='--', which='major')
plt.show()
```

計算定積分 $\int_{-2}^6 \sqrt{x^2+4x+12}\;dx$ 並繪製所代表的不規則面積

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integral
 
# define anonymous function
f = lambda x : (x ** 2 + 4 * x + 12) ** 0.5
lb, ub = -2, 6
result = integral.quad(f, lb, ub)
print("The integration is {:.4f}".format(result[0]))
 
# draw the function line -------------------
x = np.linspace(-5, 20, 100)
y = f(x)
fig, ax = plt.subplots(1, 1) 
ax.plot(x, y)
# draw the area ----------------------------
x = np.linspace(lb, ub, 100)
y = f(x)
# fill color between y and 0
ax.fill_between(x, y, 0, alpha=0.3, color='b')
ax.set_xlabel('X')
ax.set_title("The area = {:.4f}".format(result[0]))
plt.grid(True, linestyle='--', which='major')
plt.xlim(-5,10)
plt.ylim(0,10)
plt.show()
```

$F(x) = \int_{-\infty}^{x} \frac{1}{\sqrt{2\pi}} e^{-t^2/2}\; dt for -5 \leq x \leq 5.$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integral

g = lambda x: np.exp(-x**2/2)/np.sqrt(2*np.pi)
# f = lambda x: integral.quad(g , -100, x)
x = np.linspace(-5, 5, 10000)
F = np.zeros(len(x))

for i in range(len(x)):
    F[i] = integral.quad(g, -100, x[i])[0]


fig,ax = plt.subplots(1,1)
ax.plot(x,F)
plt.grid(True, linestyle='--', which='major')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integral
 
n = 10000
f = lambda x : np.exp(-x**2/2) / np.sqrt(2*np.pi)
xlim = [-5, 5]
x = np.linspace(xlim[0], xlim[1], n)
 
def cdf(x):
    return integral.quad(f, -np.inf, x)[0]
 
vec_F = np.vectorize(cdf) # define a vectorized function
F = vec_F(x) # evalaute cdf for all x
plt.plot(x, F, drawstyle='steps-pre') # stairs plot
plt.title('A stair plot of CDF function by integration')
plt.show()
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
 
# print(norm.__doc__) # see basic information
mu, sigma = 0, 1
xlim = [mu - 5 * sigma, mu + 5 * sigma]
x = np.linspace(xlim[0], xlim[1], 1000)
y = norm.pdf(x, mu, sigma)
 
plt.plot(x, y)
plt.style.use('fivethirtyeight') # 538 style
 
# patch the area of confidence interval
conf = 0.95
ci = norm.ppf([(1-conf)/2, (1+conf)/2]) # inverse CDF　function
x_ci = np.linspace(ci[0], ci[1], 1000)
y_ci = norm.pdf(x_ci, mu, sigma)
 
plt.fill_between(x_ci, y_ci, 0, color = 'g', alpha = 0.5)
plt.grid(True, linestyle='--', which='major')
plt.title('Normal({}, {})'.format(mu, sigma))
plt.show()
```
