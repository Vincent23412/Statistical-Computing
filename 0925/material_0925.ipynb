```python
import numpy as np
import matplotlib.pyplot as plt

# The codes below demonstrate the ticks and labels in x axis
x = np.linspace(-3*np.pi, 3*np.pi, 200)
# xx  =x.reshape(1,-1) (1,200)
y = 4 / np.pi * np.sin(x)
fig, ax = plt.subplots(1, figsize=(9,6))
ax.plot(x, y, color='b')
ax.set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi)
ax.set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0',
                   '$\pi$', '2$\pi$', '3$\pi$'],  fontsize=10, color='b')
ax.set_ylim([-3, 3])
ax.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3*np.pi, 3*np.pi, 200)
rows = 2
cols = 2
fig, axs = plt.subplots(rows, cols, figsize=(9,6))


for i, ax in enumerate(axs.flat):
    n = i + 1
    y = 0
    for k in range(n) :
        l = 2 * k + 1
        y = y + np.sin(l * x) / l
        ax.plot(x , 4 * y / np.pi)
        ax.set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3]) * np.pi)
        ax.set_xticklabels(
            ["-3$\pi$", "-2$\pi$", "$\pi$", "0", "$\pi$", "2$\pi$", "3$\pi$"],
            fontsize=10, color="b",
        )
    
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
n = [1, 2, 3]
colors = ['r', 'g', 'b']
x = np.linspace(-3*np.pi, 3*np.pi, 200)
for i in n:         
    y=np.sin(i*x)   
    plt.plot(x, y, label = 'sin({}x)'.format(i), \
        color = colors[i - 1])
 
plt.legend()
plt.ylim(-2.5, 2.5), plt.grid(True), plt.xlabel('X')
plt.title('Demonstration of legend for each line')
plt.show()
```
