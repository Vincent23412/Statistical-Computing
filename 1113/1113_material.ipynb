```python
from scipy.stats import norm
from scipy.stats import skew
from matplotlib import pyplot as plt
import numpy as np

n = 10
N = 10000
X = norm.rvs(size=(N,n))
g1 = np.sqrt(n / 6) * skew(X,axis=1)
plt.hist(g1, bins=50, density=True, label='g1')

xx = np.linspace(-4,4,1000)
plt.plot(xx, norm.pdf(xx), label='N(0,1)')
plt.show()

```

```python
from scipy.stats import chi2
import numpy as np
from scipy.stats import skew, t,norm

n = 100
N = 50000
alpha = 0.05
CV = norm.ppf(1-alpha/2)

# x = chi2.rvs(50, size=(N,n))
# x = norm.rvs(size=(N,n))
x = t.rvs(1,size=(N,n))

g1 = np.sqrt(n/6) * skew(x, axis=1)
reject = np.abs(g1) > CV
reject_rate = np.mean(reject)  # power
print(reject_rate)
# plt.hist(g1, bins=50, density=True, label='g1')
# plt.show()
```

```python
# kurtosiskurtosis

from scipy.stats import norm, kurtosis
from scipy.stats import skew
from matplotlib import pyplot as plt
import numpy as np

n = 2000
N = 10000
X = norm.rvs(size=(N,n))
g2 = np.sqrt(n/24) * kurtosis(X,bias=True, axis=1)
plt.hist(g2, bins=50, density=True, label='g1')

xx = np.linspace(-4,4,1000)
plt.plot(xx, norm.pdf(xx), label='N(0,1)')
plt.show()

```

```python
# g3 = g1 + g2

from scipy.stats import norm
from scipy.stats import skew
from matplotlib import pyplot as plt
import numpy as np

n = 10
N = 10000
X = norm.rvs(size=(N,n))
g1 = np.sqrt(n / 6) * skew(X,axis=1)
g2 = np.sqrt(n/24) * kurtosis(X,bias=True, axis=1)
g3 = g1 + g2
plt.hist(g3, bins=50, density=True, label='g1')

xx = np.linspace(-4,4,1000)
plt.plot(xx, norm.pdf(xx), label='N(0,1)')
plt.show()

```
